```python
import os

folder_path = 'data/teamA'

# 모든 .md 파일 이름을 하나의 리스트로 통합
all_md_files = []

for folder_name in os.listdir(folder_path):
    folder_full_path = os.path.join(folder_path, folder_name)
    if os.path.isdir(folder_full_path):
        files = os.listdir(folder_full_path)
        all_md_files.extend([f for f in files if f.endswith('.md')])

print(all_md_files)
```

```python
import pandas as pd
import os

# CSV 파일을 읽어들임
csv_file_path = 'document/url_table.csv'
url_table = pd.read_csv(csv_file_path)

# 실제 .md 파일 이름 리스트를 생성함
# 여기서는 실제 파일 시스템에서 파일 이름을 가져와야 함
md_files_path = 'data/teamA/'

# CSV 파일의 'source' 컬럼과 .md 파일 이름 리스트의 길이 비교
if len(url_table['source']) == len(all_md_files):
    # 길이가 같다면 'source' 컬럼을 새 파일 이름으로 업데이트
    url_table['source'] = all_md_files
    # 변경된 데이터프레임을 새로운 CSV 파일로 저장
    new_csv_file_path = 'document/url_table_team_a.csv'
    url_table.to_csv(new_csv_file_path, index=False)
    print(f"CSV file updated and saved to {new_csv_file_path}")
else:
    # 길이가 다르다면 오류 메시지 출력
    print("The length of the 'source' column does not match the length of the .md files list.")
```
