```python
## 합치기
import os

filepath = "temp_gen"
combined_csv = pd.DataFrame()

for file_name in os.listdir(filepath) :
    if file_name.endswith(".csv"):
        csv = pd.read_csv(os.path.join(f"{filepath}/{file_name}"), encoding="utf-8")
        
        ### check data sanity "Documents" column.
        if csv["Documents"].nunique() != 1:
            print(file_name)

        combined_csv = pd.concat([combined_csv, csv], ignore_index=True)
```
